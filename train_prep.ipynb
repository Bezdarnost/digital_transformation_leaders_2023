```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
train = pd.read_csv('train_v3_pairs.csv')
```

```python
df
```

```python
train
```

```python
features = (
    train
    .merge(
        df
        .add_suffix('1'),
        on="variantid1"
    )
    .merge(
        df
        .add_suffix('2'),
        on="variantid2"
    )
)
```

```python
features.to_csv('train_v3.csv', index=False)
```
