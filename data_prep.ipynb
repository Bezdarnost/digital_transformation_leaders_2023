```python
df["cat3"] = df["categories"].apply(lambda x: json.loads(x)["3"])
cat3_counts = df["cat3"].value_counts().to_dict()

cntr = 0

for cat3 in cat3_counts:
    if cat3_counts[cat3] < 1_000:
        cntr += cat3_counts[cat3]
        
cntr
```

```python
df["cat3_grouped"] = df["cat3"].apply(lambda x: x if cat3_counts[x] > 1000 else 'rest')
```

```python
df.drop(['name', 'pic_embeddings_resnet_v1', 'characteristic_attributes_mapping', 'categories', 'cat3'], axis=1, inplace=True)
df
```

```python
print(f'Count of unique colors: {len(df.color_parsed.unique())}\nCount of unique cat3: {len(df.cat3_grouped.unique())}')
```

```python
df = pd.get_dummies(df, columns=['color_parsed', 'cat3_grouped'])
df
```

```python
df.reset_index(inplace=True, drop=True)
for i in range(128):
    nums = [df['main_pic_embeddings_resnet_v1'][j][0][i] for j in range(len(df))]
    df[f'pic{i}'] = nums
```

```python
for i in range(64):
    nums = [df['name_bert_64'][j][i] for j in range(len(df))]
    df[f'name{i}'] = nums
```
